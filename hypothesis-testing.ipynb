#### Hypothesis Testing

A research group has been investigating a new golf ball coating designed to resist cuts and provide a more durable ball.Concerns about the effect of the new coating on driving distance has been raised. To compare, 40 balls of both new and current were subjected to a distance test. 

With a significance level of 0.05 should the null hypothesis be rejected.

```python
#imports

import pandas as pd
import seaborn as sns

from scipy import stats
```

```python
data = pd.read_csv("data/Golf.csv")
```

#### Ho:
The new coat causes no effect on the driving distance of the ball

#### Ha:

The new coat causes an effect on the driving distance of the ball

```python
data.head()
```

```python
sns.histplot(data['Current'], kde=True)
```

```python
sns.histplot(data['New'], kde=True)
```

```python
stat, p_value = stats.ttest_ind(data['Current'], data['New'])

stat, p_value
```

The signifaicance level is less than the p_value (0.18793), therefore the null hypothesis is not rejected (accept the null hypothesis). *Meaning the new coating causes no effect onthe driving distance of the ball.*
